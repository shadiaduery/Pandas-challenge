```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# Calculate the total number of schools by first identifying the number of unique schools
unique_schools = school_data_complete.loc[:,"school_name"].unique()
# unique_schools_num = unique_schools.count()
# unique_schools_num
unique_schools_df = pd.DataFrame(unique_schools)
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
pass_math = school_data_complete["math_score"]>=70
pass_math_final = (pass_math.sum()/total_students)*100
# pass_math_final
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
pass_reading = school_data_complete["reading_score"]>=70
pass_reading_final = (pass_reading.sum()/total_students)*100
# pass_reading_final
```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_pass = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]
overall_pass_final = overall_pass["student_name"].count()
overall_pass_final = overall_pass_final/total_students*100
# overall_pass_final
```

```python
# Calculate the total number of students
total_students
# Calculate the total budget
total_budget
# Calculate the average math score
avg_math_scores
# Calculate the average reading score
avg_read_scores
# Calculate the percentage of students with a passing math score (70 or greater)
pass_math_final
# Calculate the percentage of students with a passing reading score (70 or greater)
pass_reading_final
# Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_pass_final
# Create a dataframe to hold the above results
results =[{
    "Total Schools":15,
    "Total Students" :total_students, 
    "Total Budget":total_budget, 
    "Average Math Scores": avg_math_scores, 
    "Average Reading Scores": avg_read_scores,
    "% Passing Math":pass_math_final, 
    "% Passing Reading":pass_reading_final, 
    "% Overall Passing":overall_pass_final,
          }]

results_df = pd.DataFrame(results)
results_clean_df = results_df
results_clean_df["Total Students"]= results_clean_df["Total Students"].map("{:,}".format)
results_clean_df["Total Budget"]= results_clean_df["Total Budget"].map("${:,}".format)
results_clean_df["Average Math Scores"]=results_clean_df["Average Math Scores"].map("{:.2f}".format)
results_clean_df["Average Reading Scores"]=results_clean_df["Average Reading Scores"].map("{:.2f}".format)
results_clean_df["% Passing Math"]=results_clean_df["% Passing Math"].map("{:.2f}".format)
results_clean_df["% Passing Reading"]=results_clean_df["% Passing Reading"].map("{:.2f}".format)
results_clean_df["% Overall Passing"]= results_clean_df["% Overall Passing"].map("{:.2f}".format)
results_clean_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_summary_df = school_data_complete.groupby("school_name")

school_grade = school_summary_df["grade"].first()
school_type = school_summary_df["type"].first()
total_students_sc = school_summary_df["Student ID"].count()
total_budget_sc = school_summary_df["budget"].first()
per_st_budg_sc = total_budget_sc/total_students_sc
avg_math_score_sc = school_summary_df["math_score"].sum()/total_students_sc
avg_read_score_sc = school_summary_df["reading_score"].sum()/total_students_sc

pass_math_sch = school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name")["math_score"].count()
pass_math_sch_final = pass_math_sch/total_students_sc*100


pass_read_sch = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name")["reading_score"].count()
pass_read_sch_final = pass_read_sch/total_students_sc*100


overall_pass = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)].groupby("school_name")
overall_pass_final = overall_pass["student_name"].count()
overall_pass_final = overall_pass_final/total_students_sc*100

results =pd.DataFrame(school_type)
results["Total Students"] = total_students_sc
results["Total Budget"] = total_budget_sc
results["Budget ($/student)"] = per_st_budg_sc
results["Average Math Scores"] = avg_math_score_sc 
results["Average Reading Scores"] = avg_read_score_sc
results["% Passing Math"] = pass_math_sch_final
results["% Passing Reading"] = pass_read_sch_final
results["% Overall Passing"] = overall_pass_final
results = results.rename (columns = {"type":"School Type"})
results_clean1_df = results 
# results_clean1_df

# important to note that the formatting down below has to be applied at the end, because some of the calculations would not be able to perform numerical calculations on strings

results_clean1_df["Total Students"]= results_clean1_df["Total Students"].map("{:,}".format)
results_clean1_df["Total Budget"]= results_clean1_df["Total Budget"].map("${:,}".format)
results_clean1_df["Budget ($/student)"]= results_clean1_df["Budget ($/student)"].map("${:.0f}".format)
results_clean1_df["Average Math Scores"]=results_clean1_df["Average Math Scores"].map("{:.2f}".format)
results_clean1_df["Average Reading Scores"]=results_clean1_df["Average Reading Scores"].map("{:.2f}".format)
results_clean1_df["% Passing Math"]=results_clean1_df["% Passing Math"].map("{:.2f}".format)
results_clean1_df["% Passing Reading"]=results_clean1_df["% Passing Reading"].map("{:.2f}".format)
results_clean1_df["% Overall Passing"]= results_clean1_df["% Overall Passing"].map("{:.2f}".format)
results_clean1_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
student_spending_schools_df = results[["Budget ($/student)", "Average Math Scores", "Average Reading Scores", "% Passing Math","% Passing Reading","% Overall Passing"]]
student_spending_schools_df = student_spending_schools_df.sort_values("Budget ($/student)", ascending=False)
student_spending_schools_df

# Create the bins in which Data will be held  
bins = [555, 580, 605, 630, 655]

# Create the names for the five bins
group_names = ["555-579", "580-604", "605-629", "630-655"]

student_spending_schools_df["Budget ($/student)"] = pd.cut(student_spending_schools_df["Budget ($/student)"], bins, labels=group_names, include_lowest=True)
student_spending_schools_df

# Creating a group based off of the bins
student_spending_schools_df = student_spending_schools_df.groupby("Budget ($/student)")
student_spending_schools_df.mean()

```
