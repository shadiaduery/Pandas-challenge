```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_summary_df = school_data_complete.groupby("school_name")

school_grade = school_summary_df["grade"].first()
school_type = school_summary_df["type"].first()
total_students_sc = school_summary_df["Student ID"].count()
total_budget_sc = school_summary_df["budget"].first()
per_st_budg_sc = total_budget/total_students_sc
avg_math_score_sc = school_summary_df["math_score"].sum()/total_students_sc
avg_read_score_sc = school_summary_df["reading_score"].sum()/total_students_sc

pass_math_sch = school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name")["math_score"].count()
pass_math_sch_final = pass_math_sch/total_students_sc*100


pass_read_sch = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name")["reading_score"].count()
pass_read_sch_final = pass_read_sch/total_students_sc*100


overall_pass = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)].groupby("school_name")
overall_pass_final = overall_pass["student_name"].count()
overall_pass_final = overall_pass_final/total_students_sc*100

results =pd.DataFrame(school_type)
results["Total Students"] = total_students_sc
results["Total Budget"] = total_budget_sc
results["Budget ($/student)"] = per_st_budg_sc
results["Avg Math Scores"] = avg_math_score_sc 
results["Avg Reading Scores"] = avg_read_score_sc
results["Pass Math"] = pass_math_sch_final
results["Pass Reading"] = pass_read_sch_final
results["Overall Pass"] = overall_pass_final
results = results.rename (columns = {"type":"School Type","school_name": "School Name"})
results
```

```python
top_schools_df = results.sort_values("Overall Pass", ascending=False)
top_schools_df.head(5)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
schools_grade = school_data_complete[["school_name","grade", "math_score"]]
schools_grades = school_data_complete[["school_name", "math_score"]]

grade_9th = schools_grade.loc[schools_grade["grade"] =="9th",:]
grade_9th_summary = grade_9th.groupby("school_name").mean()


grade_10th = schools_grade.loc[schools_grade["grade"] =="10th",:]
grade_10th_summary = grade_10th.groupby("school_name").mean()

grade_11th = schools_grade.loc[schools_grade["grade"] =="11th",:]
grade_11th_summary = grade_11th.groupby("school_name").mean()

grade_12th = schools_grade.loc[schools_grade["grade"] =="12th",:]
grade_12th_summary = grade_12th.groupby("school_name").mean()

grade_9th_summary ["10th grade"] =grade_10th_summary
grade_9th_summary ["11th grade"] =grade_11th_summary
grade_9th_summary ["12th grade"] =grade_12th_summary
grade_9th_summary = grade_9th_summary.rename(columns = {"math_score":"9th grade"})
grade_9th_summary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
student_spending_schools_df = results[["Budget ($/student)", "Avg Math Scores", "Avg Reading Scores", "Pass Math","Pass Reading","Overall Pass"]]
student_spending_schools_df = student_spending_schools_df.sort_values("Budget ($/student)", ascending=False)
student_spending_schools_df

# Create the bins in which Data will be held  
bins = [5, 15, 25, 50, 95]

# Create the names for the five bins
group_names = ["5-15", "15-25", "25-50", "50-95"]

student_spending_schools_df["Budget ($/student)"] = pd.cut(student_spending_schools_df["Budget ($/student)"], bins, labels=group_names, include_lowest=True)
student_spending_schools_df

# Creating a group based off of the bins
student_spending_schools_df = student_spending_schools_df.groupby("Budget ($/student)")
student_spending_schools_df.max()
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
size_schools_df = results[["Total Students", "Avg Math Scores", "Avg Reading Scores", "Pass Math","Pass Reading","Overall Pass"]]
size_schools_df = size_schools_df.sort_values("Total Students", ascending=False)
size_schools_df

# Create the bins in which Data will be held  
bins = [1000, 2000, 3000, 4000, 5000]

# Create the names for the five bins
group_names = ["1-2k", "2-3k", "3-4k", "4-5k"]

size_schools_df["School Size (# students)"] = pd.cut(size_schools_df["Total Students"], bins, labels=group_names, include_lowest=True)
size_schools_df

# Creating a group based off of the bins
size_schools_df = size_schools_df.groupby("School Size (# students)")
size_schools_df.max()
```
